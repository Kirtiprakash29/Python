Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
Note: Also add title, xlabel, and ylabel to the plot.

Scatter plots are used to observe relationship between variables and uses dots to represent the relationship between them. The scatter() method in the matplotlib library is used to draw a scatter plot. Scatter plots are widely used to represent relation among variables and how change in one affects the other.

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
print(x)
print(y)
len(x)
import matplotlib.pyplot as plt
z = plt.scatter(x,y,c= 'r')
plt.xlabel("Random numbers b/w 0 to 2 with 3 added")
plt.ylabel("Random numbers b/w 0 to 2 with 3 added with respect to length")
plt.title("SCATTER PLOT")
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

A bar plot or bar chart is a graph that represents the category of data with rectangular bars with lengths and heights that is proportional to the values which they represent. The bar plots can be plotted horizontally or vertically. A bar chart describes the comparisons between the discrete categories.

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.figure(figsize=(12,3))
plt.subplot(1,2,1)
plt.bar(company,profit)
plt.xlabel("Company Names")
plt.ylabel("Profit Incurred")
plt.title("Bar plot of profit of various IT giants")

plt.subplot(1,2,2)
plt.barh(company,profit)
plt.ylabel("Company Names")
plt.xlabel("Profit Incurred")
plt.title("Bar plot of profit of various IT giants")
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

A Box Plot is also known as Whisker plot is created to display the summary of the set of data values having properties like minimum, first quartile, median, third quartile and maximum. In the box plot, a box is created from the first quartile to the third quartile, a vertical line is also there which goes through the box at the median. Here x-axis denotes the data to be plotted while the y-axis shows the frequency distribution.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
plt.subplot(1,2,1)
plt.boxplot(box1)
plt.subplot(1,2,2)
plt.boxplot(box2)
plt.show()
```
